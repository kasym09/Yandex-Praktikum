#### 1.2. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ. 

```python
#—Å–æ–∑–¥–∞—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è:
def recovery_calc(rougher_output, rougher_input, rougher_tail):
    recovery = (rougher_output * (rougher_input - rougher_tail)) \
    / (rougher_input * (rougher_output - rougher_tail)) * 100
    return recovery
```

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚úîÔ∏è\
<font color='green'>–†–∞—Å—á—ë—Ç –≤–µ—Ä–Ω—ã–π!</font>

```python
#–≤—ã—á–∏—Å–ª—è—é —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery:
recovery = recovery_calc(train['rougher.output.concentrate_au'],
                         train['rougher.input.feed_au'],
                         train['rougher.output.tail_au'])
recovery
```

```python
#–Ω–∞—Ö–æ–∂—É MAE –º–µ–∂–¥—É —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery:
mae = mean_absolute_error(train['rougher.output.recovery'], recovery)
print('MAE –º–µ–∂–¥—É —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery =', mae)
```

## –ú–æ–¥–µ–ª—å

#### 3.1. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE.

```python
#—Å–æ–∑–¥–∞—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è sMAPE:
def sMAPE_calc(y_target, y_predict):
    smape = np.mean(abs(y_target - y_predict) / ((abs(y_target) + abs(y_predict)) / 2)) * 100
    return smape
```

```python
#—Å–æ–∑–¥–∞—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ sMAPE:
def final_sMAPE_calc(smape_rougher, smape_final):
    return (0.25*smape_rougher + 0.75*smape_final)
```

<font color='green'> `sMAPE_calc` –∏ `final_sMAPE_calc` —Ä–∞—Å—á–∏—Ç–∞–Ω—ã –≤–µ—Ä–Ω–æ.</font>

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ : </b></font> ‚úîÔ∏è <br>

```python
model_rfr = make_pipeline(StandardScaler(), RandomForestRegressor())

model_dtr = make_pipeline(StandardScaler(), DecisionTreeRegressor())

model_lr = make_pipeline(StandardScaler(), LinearRegression())
```

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚úîÔ∏è\
<font color='green'>–ö—Ä—É—Ç–æ! Pipeline –ø–æ–∑–≤–æ–ª—è–µ—Ç –∏–∑–±–µ–∂–∞—Ç—å —É—Ç–µ—á–µ–∫ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ —á–µ—Ä–µ–∑ –°–∫–∞–ª–µ—Ä –¥–∞–∂–µ –ø—Ä–∏ –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏–∏.</font>

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚ö†Ô∏è\
<font color='darkorange'> –î–ª—è –∫–∞–∫–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω—É–∂–Ω–∞ –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è?</font>

```python
#–ø–æ–∏—Å–∫ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_rougher_rfr = GridSearchCV(model_rfr, param_grid=parameters_rfr, cv=5, scoring=score_smape)
grid_final_rfr = GridSearchCV(model_rfr, param_grid=parameters_rfr, cv=5, scoring=score_smape)

grid_rougher_dtr = GridSearchCV(model_dtr, param_grid=parameters_dtr, cv=5, scoring=score_smape)
grid_final_dtr = GridSearchCV(model_dtr, param_grid=parameters_dtr, cv=5, scoring=score_smape)

grid_rougher_lr = GridSearchCV(model_lr, param_grid=parameters_lr, scoring=score_smape)
grid_final_lr = GridSearchCV(model_lr, param_grid=parameters_lr, scoring=score_smape)
```
